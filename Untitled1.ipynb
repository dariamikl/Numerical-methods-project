```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x0 = 1
y0 = 1
xb = 2.3
h = 0.1
n = int((xb-x0)/h)
```

```python
x = np.linspace(x0, xb, n)
x
```

```python
y = np.zeros([n])
```

```python
y[0] = y0
for i in range (1,n):
    y[i] = y[i-1] + h*(y[i-1]/x[i-1] + x[i-1]/y[i-1])
for i in range (n):
    print(f'{i}: {x[i]}: {y[i]}')
```

```python
%matplotlib notebook
plt.plot(x,y,'o')
plt.xlabel('X value')
plt.ylabel('Y label')
plt.title("Euler method")
plt.show
```
